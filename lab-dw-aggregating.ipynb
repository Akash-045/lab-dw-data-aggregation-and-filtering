```python
import pandas as pd
# Load data
df = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv")
#df.head()
#display(df.columns)(column names)
#df.info()(basic info)
#print(df['Total Claim Amount'].describe())(Summary statistics)
```

```python
# Filtering
low_claim_amount = df[df['Total Claim Amount'] < 1000]#(create a boolean condition)

low_claim_amount
#display(low_claim_amount.shape)
```

```python
# Have a response: Yes
response_yes = df[df['Response'] == 'Yes']

response_yes.head()
print(response_yes.shape)
```

```python
response_yes = df[df['Response'] == 'Yes']

# Group by policy type and gender and calculate average
group_df = response_yes.groupby(['Policy Type', 'Gender'])['Customer Lifetime Value'].mean()
group_df
```

```python
# Number of customer in each state
state_counts = df.groupby('State').size()#(.size counts the number of rows(customers) in each state)
state_counts
```

```python
# Filter states with more than 500 customers
filtered_states = state_counts[state_counts > 500]
filtered_states
```

```python
# Calculate maximum, minimum and median customer life time value by Education and Gender
min_max_median = df.groupby(["Education", "Gender"])["Customer Lifetime Value"].agg(['max','min','median'])
min_max_median
```
